```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
file_path = "titanic.csv"  # Path of dataset
data = pd.read_csv(file_path)
```

```python
data
```

```python

print(data.isnull().sum())
```

```python
# Fill missing Age with median
data['Age'].fillna(data['Age'].median(), inplace=True)

# Fill missing Embarked with mode
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
```

```python
data
```

```python
data = data.drop(columns=['Cabin','Name','Ticket','PassengerId'])
```

```python
data
```

```python
data['Sex'] = data['Sex'].map({'male': 0, 'female': 1})
data['Embarked'] = data['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
```

```python
print(data.dtypes)
print(data.isnull().sum())
```

```python
# --- a. Descriptive Statistics and Information ---

print("\n--- a. describe() ---")
print(data.describe())
print("\nObservation for describe():")
print("- 'Age', 'SibSp', 'Parch', and 'Fare' have numerical spreads. 'Age' has missing values (count is less than total entries).")
print("- 'Survived', 'Pclass', 'SibSp', and 'Parch' are treated as numerical here, but their nature is likely categorical or ordinal.")
print("- The average fare paid was around 32, with a significant standard deviation indicating a wide range of fares.")
```

```python
print("\n--- a. info() ---")
data.info()
print("\nObservation for info():")
print("- The dataset has 891 entries and 12 columns.")
print("- 'Age', 'Cabin', and 'Embarked' columns have missing values.")
print("- The data types are a mix of integer, float, and object (string).")
print("- 'Name', 'Sex', 'Ticket', 'Cabin', and 'Embarked' are categorical features stored as objects.")
```

```python

# Heatmap for correlations
plt.figure(figsize=(10, 6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Pairplot for key variables
# Selecting numerical columns for pairplot

sns.pairplot(data, hue='Survived')
plt.suptitle("Pairplot of Numerical Features (Hue: Survived)", y=1.02)
plt.show()
print("\nObservation for pairplot():")
print("- There might be some separation in 'Fare' distribution based on 'Survived'.")
print("- Passengers in higher 'Pclass' (lower numerical value) seem to have a slightly higher survival rate.")
print("- The relationship between 'Age', 'SibSp', 'Parch' and 'Survived' is not immediately obvious from these pairwise scatter plots.")
```

```python
# Set up the seaborn theme
sns.set(style='whitegrid')

# Survival Count Plot
plt.figure(figsize=(6, 4))
sns.countplot(data, x='Survived', palette='pastel')
plt.xticks([0, 1], ['Did Not Survive', 'Survived'])
plt.title('Overall Survival Count')
plt.ylabel('Number of Passengers')
plt.xlabel('Survival Status')
plt.show()

# Survival Rate
survival_rate = data['Survived'].mean() * 100
print(f"Overall Survival Rate: {survival_rate:.2f}%")
```

```python
# Survival rate by class
sns.barplot(data, x='Pclass', y='Survived')
plt.title("Survival Rate by Passenger Class")
plt.show()
```

```python
# Survival rate by gender
sns.barplot(data, x='Sex', y='Survived')
plt.title("Survival Rate by Gender")
plt.show()
```

```python
# Age distribution by survival
sns.histplot(data, x='Age', hue='Survived', kde=True, bins=30)
plt.title("Age Distribution by Survival")
plt.show()
```

```python
sns.histplot(data['Fare'], bins=40, kde=True)
plt.title("Fare Distribution")
plt.show()
```

```python
sns.boxplot(data, x='Survived', y='Age')
plt.title("Age Distribution by Survival")
plt.show()
```

```python
sns.scatterplot(data, x='Age', y='Fare', hue='Survived')
plt.title("Fare vs Age by Survival")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Pclass', hue='Survived',data=data)
plt.title("Survival Count by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers")
plt.xticks([0, 1, 2], ['1st', '2nd', '3rd'])
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
print("\nObservation for Pclass vs. Survival:")
print("- Passengers in the 1st class had a significantly higher survival rate compared to the 2nd and 3rd classes.")
print("- The 3rd class had the lowest survival rate.")
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Sex', hue='Survived', data=data)
plt.title("Survival Count by Sex")
plt.xlabel("Sex")
plt.ylabel("Number of Passengers")
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
print("\nObservation for Sex vs. Survival:")
print("- Female passengers had a much higher survival rate than male passengers.")
```

```python
# Countplot of Survival by Pclass
plt.figure(figsize=(6, 4))
sns.countplot(data, x='Pclass', hue='Survived', palette='Set2')
plt.title('Survival by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()

# Survival rate per class
class_survival = data.groupby('Pclass')['Survived'].mean()
print("Survival Rate by Passenger Class:")
print(class_survival)
```

```python
# Fare Distribution
plt.figure(figsize=(8, 4))
sns.histplot(data, x='Fare', bins=40, kde=True, color='orchid')
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Number of Passengers')
plt.xlim(0, 300)  # Limit axis to better visualize distribution
plt.show()

# Boxplot: Fare by Survival
plt.figure(figsize=(6, 4))
sns.boxplot(data, x='Survived', y='Fare', palette='magma')
plt.xticks([0, 1], ['Did Not Survive', 'Survived'])
plt.title('Fare vs Survival')
plt.show()

# Survival rate vs fare brackets
data['FareRange'] = pd.cut(data['Fare'], bins=[0, 50, 100, 200, 600], labels=['<50', '50-100', '100-200', '200+'])
fare_survival = data.groupby('FareRange')['Survived'].mean()
print("Survival Rate by Fare Range:")
print(fare_survival)
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Embarked', hue='Survived', data=data)
plt.title("Survival Count by Port of Embarkation")
plt.xlabel("Port of Embarkation")
plt.ylabel("Number of Passengers")
plt.xticks([0, 1, 2], ['Cherbourg (C)', 'Queenstown (Q)', 'Southampton (S)'])
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
print("\nObservation for Embarked vs. Survival:")
print("- Passengers who embarked from Cherbourg (C) had a relatively higher survival rate compared to those from Southampton (S) and Queenstown (Q).")
```
